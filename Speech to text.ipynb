# Changing .mp3 to .wav format

```python
# Source directory with .mp3 files
source_directory = r"E:\AudioPalm_MODEL\Voices"

# Destination directory to save .wav files
destination_directory = r"E:\AudioPalm_MODEL\formatted_voices"

# Create the destination directory if it does not exist
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)

# Get a list of all files in the source directory
file_list = os.listdir(source_directory)


# Convert .mp3 files to .wav and save in the destination directory
for filename in file_list:
    if filename.lower().endswith(".mp3"):
        mp3_path = os.path.join(source_directory, filename)
        print(mp3_path)
        # Load the mp3 file
        audio = pydub.AudioSegment.from_mp3(mp3_path)
        wav_filename = os.path.splitext(filename)[0] + ".wav"
        wav_path = os.path.join(destination_directory, wav_filename)
        print(wav_path)
        
        
        # Export the audio as wav
        audio.export(wav_path, format="wav")

        print(f"Converted {filename} to {wav_filename} and saved in {destination_directory}.")
```
